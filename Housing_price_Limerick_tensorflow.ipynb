Try to predict housing price in Limerick City 

Xs = number of rooms
Ys = number of baths
Zs = price based on Daft.ie

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
xs = np.array([2.0, 3.0, 2.0,3.0,1.0,2.0], dtype=float)
ys = np.array([2.0, 1.0, 1.0,3.0,1.0,2.0], dtype=float)
Zs = np.array([90000.0, 160000.0, 110000.0,165000.0,79000.0,157000.0], dtype=float)
model.fit(xs, zs, epochs=500)
```

```
print(model.predict([7.0]))
```
